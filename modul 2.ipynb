```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('C:/Users/panca/OneDrive/Documents/praktikum_pcd/modul2/scrambled image.jpg', cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

height, width, _ = image.shape
fig, ax = plt.subplots(1)

ax.imshow(image)

ax.set_aspect('equal'
ticks = np.arange(0, max(width, height), max(width, height) // 20)

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels((ticks * 0.264583).astype(int), fontsize=8)  # Increase the font size of the x tick labels
ax.set_yticklabels((ticks * 0.264583).astype(int), fontsize=8)  # Increase the font size of the y tick labels

ax.set_xlabel('Millimeters', fontsize=10)  # Increase the font size of the x label
ax.set_ylabel('Millimeters', fontsize=10)  # Increase the font size of the y label

plt.show()
```
